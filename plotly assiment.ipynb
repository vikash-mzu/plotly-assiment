```python
Q1. Load the "titanic dataset using Seaborn and plot a scatter plot for age and fare columns using Plotly Express.

import seaborn as sns
import plotly.express as px

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Create a scatter plot for 'age' and 'fare'
fig = px.scatter(titanic, x='age', y='fare', title='Age vs Fare Scatter Plot')

# Show the plot
fig.show()



Q2. Using the tips dataset in Plotly Express, plot a box plot.

import plotly.express as px

# Load the tips dataset
tips = px.data.tips()

# Create a box plot for total_bill by day
fig = px.box(tips, x='day', y='total_bill', title='Box Plot of Total Bill by Day')

# Show the plot
fig.show()



Q3. Plot a histogram for x = "sex" and y = "total_bill", use "smoker" with pattern_shape, and "day" with color.

import plotly.express as px

# Load the tips dataset
tips = px.data.tips()

# Create a histogram with specified parameters
fig = px.histogram(tips, x="sex", y="total_bill", color="day", pattern_shape="smoker",
                   title="Total Bill by Sex, Smoker, and Day")

# Show the plot
fig.show()



Q4. Using the iris dataset in Plotly, plot a scatter matrix using "species" for the color parameter.

import plotly.express as px

# Load the iris dataset
iris = px.data.iris()

# Create a scatter matrix for the specified columns and species as color
fig = px.scatter_matrix(iris, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"], 
                        color="species", title="Scatter Matrix of Iris Dataset")

# Show the plot
fig.show()


Q5. What is Distplot? Plot a distplot using Plotly Express.
Distplot Explanation: A distplot (distribution plot) shows the distribution of a univariate set of observations, combining a histogram with a kernel density estimate or other representations of the data.

python
Copy code
import plotly.express as px

# Load the tips dataset
tips = px.data.tips()

# Plot a distplot (represented by a histogram and density estimate)
fig = px.histogram(tips, x="total_bill", nbins=30, marginal="violin", title='Distplot of Total Bill')

# Show the plot
fig.show()
```
